# Exercise 2: Vegetation Index Analysis
In this exercise, you will analyze vegetation changes within one growing season in a selected agricultural region (Kujawy, Poland).

## Task 1: Define AOI
Create an Area of Interest (AOI) using the coordinates provided below.

```python
import ee
ee.Authenticate()
ee.Initialize(project='ee-kkosciukk')
```

```python
# TODO: Define AOI
aoi = ee.Geometry.Rectangle([19.0, 53.5, 19.5, 54.0])
```

## Task 2: Retrieve Sentinel-2 Imagery
Retrieve two Sentinel-2 satellite images from the following dates: 2023-04-01 and 2023-07-01.

```python
# TODO: Retrieve Sentinel-2 images for the selected dates.
sentinel2 = ee.ImageCollection('COPERNICUS/S2_HARMONIZED') \
    .filterDate('2023-04-01', '2023-04-30') \
    .filterBounds(aoi)\
    .sort('CLOUDY_PIXEL_PERCENTAGE')  # Sort by cloud cover percentage

#the least cloudy image
image1 = sentinel2.first()

#metadata of the selected image
metadata1= image1.getInfo()
metadata1
```

```python
# TODO: Retrieve Sentinel-2 images for the selected dates.
sentinel22= ee.ImageCollection('COPERNICUS/S2_HARMONIZED') \
    .filterDate('2023-07-01', '2023-07-31') \
    .filterBounds(aoi) \
    .sort('CLOUDY_PIXEL_PERCENTAGE') 

#the least cloudy image
image2 = sentinel22.first()

#metadata of the selected image
metadata2= image2.getInfo()
metadata2
```

## Task 3: Calculate NDVI
Calculate the Normalized Difference Vegetation Index (NDVI) for both images.

```python
# TODO: Calculate NDVI for both images (Choose the appropriate S-2 bands).
image1 = image1.clip(aoi)
image2= image2.clip(aoi)
```

```python
# NDVI for the first image
ndvi1 = image1.normalizedDifference(['B8', 'B4']).rename('NDVI1')

# NDVI for the second image
ndvi2 = image2.normalizedDifference(['B8', 'B4']).rename('NDVI2')


print(ndvi1.getInfo())
print(ndvi2.getInfo())
```

## Task 4: Compare NDVI
Display both NDVI indices on the map and calculate and visualize their difference.

```python
# TODO: Compare NDVI, visualize the differences, and analyze vegetation changes.
import geemap


ndvi_difference = ndvi2.subtract(ndvi1).rename('NDVI_Difference')


ndvi_diff_viz = {
    'min': -1,
    'max': 1,
    'palette': ['red', 'white', 'green']  # Red for decrease, green for increase
}


Map = geemap.Map()
Map.centerObject(aoi, 10)
Map.addLayer(ndvi_difference, ndvi_diff_viz, 'NDVI Difference')
Map.addLayer(ndvi1, {'min': 0, 'max': 1, 'palette': ['white', 'green']}, 'NDVI April')
Map.addLayer(ndvi2, {'min': 0, 'max': 1, 'palette': ['white', 'green']}, 'NDVI July')
Map.addLayer(aoi, {}, 'AOI')
Map
```
